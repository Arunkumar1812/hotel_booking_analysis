```python
df_clean = df.copy()
```

```python
df_clean.drop(columns='company', inplace=True)
```

```python
df_clean.columns
```

```python
df_clean['children'] = df_clean['children'].fillna(0)
df_clean['country'] = df_clean['country'].fillna(df_clean['country'].mode()[0])
df_clean['agent'] = df_clean['agent'].fillna(0)
```

```python
df_clean.isnull().sum()
```

```python
df_clean[df_clean.duplicated()].head(10)
```

```python
print("Before removing duplicates:", df_clean.shape)

df_clean.drop_duplicates(inplace=True)

print("After removing duplicates:", df_clean.shape)
```

```python
df_clean['reservation_status_date'] = pd.to_datetime(
    df_clean['reservation_status_date']
)
```

```python
df_clean['reservation_status_date'].dtype
```

```python
object_cols = df_clean.select_dtypes(include='object').columns

for col in object_cols:
    df_clean[col] = df_clean[col].str.strip()
```

```python
for col in object_cols:
    print(f"\n{col}")
    print(df_clean[col].unique())
```

```python
df_clean.describe().T
```

```python
df_clean[
    (df_clean['adults'] == 0) &
    (df_clean['children'] == 0) &
    (df_clean['babies'] == 0)
]
```

```python
(
    (df_clean['adults'] == 0) &
    (df_clean['children'] == 0) &
    (df_clean['babies'] == 0)
).sum()
```

```python
df_clean['total_guests'] = (
    df_clean['adults'] +
    df_clean['children'] +
    df_clean['babies']
)
```

```python
df_clean['total_nights'] = (
    df_clean['stays_in_weekend_nights'] +
    df_clean['stays_in_week_nights']
)
```

```python
df_clean['total_revenue'] = (
    df_clean['adr'] *
    df_clean['total_nights']
)
```

```python
df_clean['booking_status'] = df_clean['is_canceled'].map({
    0: 'Completed',
    1: 'Canceled'
})
```
